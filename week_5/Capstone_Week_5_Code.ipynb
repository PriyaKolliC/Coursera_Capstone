<b>3. Reading and Scapping the Postal Code data of Toronto from the wikipedia page</b>

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
tables = pd.read_html(url, header=0)
wiki_table = tables[0]
wiki_table
```

<b>4. Filtering the above records such that "Borough" is assigned</b>

```python
wiki_table = wiki_table.loc[wiki_table['Borough'] != 'Not assigned']
wiki_table
```

```python
for row in wiki_table.iterrows():
    neighborhood = row[1][2]
    borough = row[1][1]
    if neighborhood == 'Not assigned' and borough != 'Not assigned':
        row[1][2] = borough
wiki_table
```

```python
wiki_table.describe()
```

<b>5. The instructor has provided with the choice of obtaining location coordinates either from Geocoder API or from the csv file. I chose the latter. </b>

```python
coordinates_data = pd.read_csv('Geospatial_Coordinates.csv')
coordinates_data
```

<b>6.Converting the dataframe containing location coordinates into a dictionary to easily access,  using Postal Code as the key</b>

```python
coordinates_dict = coordinates_data.set_index('Postal Code').T.to_dict('list')
coordinates_dict
```

<b>7. For each record in the scrapped data we obtain latitude and longitude values from the dictionary above </b>

```python
result_df = pd.DataFrame(columns = ["Postal Code", "Borough", "Neighborhood", "Latitude", "Longitude"])
for row in wiki_table.iterrows():
    pc = row[1][0]
    bo = row[1][1]
    nh = row[1][2]
    lat, long = coordinates_dict[pc]
    result_df = result_df.append(pd.Series([pc, bo, nh, lat, long], index = result_df.columns), ignore_index = True)
result_df
```

```python
radius = 700 
LIMIT = 100
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    latitude, 
   longitude, 
    radius, 
   LIMIT)
results = requests.get(url).json()
```

```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues)
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
filtered_columns = ['venue.name','venue.categories','venue.location.lat','venue.location.lng']
nearby_venues = nearby_venues.loc[:, filtered_columns]
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues
```

<b>9.a.e) For simplicity, looking at the top 10 venues in Scarborough borough. Note that these venues do not describe the characteristics of neighborhoods in Scarborough borough</b>

```python
top_10_borough_venue_categories = pd.Series(nearby_venues.categories)
top_10_borough_venue_categories.value_counts()[:10]
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=700):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):    
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        # making GET request
        venue_results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name'],
            v['venue']['location']['lat'], 
            v['venue']['location']['lng']
            ) for v in venue_results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Category',
                  'Venue Latitude', 
                  'Venue Longitude']
    
    return(nearby_venues)
```

<b>9.a.f) Here comes the most important part of the project - Venues of each neighborhood in the radius of 700 units</b>

```python
borough_venues = getNearbyVenues(names = borough_df['Neighborhood'],
                                   latitudes = borough_df['Latitude'],
                                   longitudes = borough_df['Longitude']
                                  )
```

```python
borough_venues
```

```python
print('There are {} Uniques Categories.'.format(len(borough_venues['Venue Category'].unique())))
borough_venues.groupby('Neighborhood').count().head()
```

```python
borough_onehot = pd.get_dummies(borough_venues[['Venue Category']], prefix="", prefix_sep="")


# add neighborhood column back to dataframe
borough_onehot['Neighborhood'] = borough_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [borough_onehot.columns[-1]] + list(borough_onehot.columns[:-1])

borough_onehot = borough_onehot[fixed_columns]
borough_grouped = borough_onehot.groupby('Neighborhood').sum().reset_index()
borough_grouped
```

<b>9.a.g) Look at the results below showing most important venues of each neighborhood along with their frequency</b>

```python
i = 1
num_top_venues = 10
for hood in borough_grouped['Neighborhood']:
    print(str(i) + ". Neighborhood: "+hood)
    temp = borough_grouped[borough_grouped['Neighborhood'] == hood].T.reset_index()
    
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
    i = i + 1
    
```

<b>9.a.h) Based on the frequency of the venues for each neighborhood, below is the more clearer view of data for each neighborhood</b>

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10
indicators = ['st', 'nd', 'rd']
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = borough_grouped['Neighborhood']

for ind in np.arange(borough_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(borough_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

<b>9.a.i) Obtaining k-means clustering approach to obtain clusters of similar neighborhoods and based on what features</b>

```python
# Using K-Means to cluster neighborhood into 5 clusters
borough_grouped_clustering = borough_grouped.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters = 5, random_state = 0).fit(borough_grouped_clustering)
kmeans.labels_
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
borough_merged = borough_df.iloc[:,:]
borough_merged = borough_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
borough_merged
```

```python
kclusters = 5
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
colors_array = cm.rainbow(np.linspace(0, 1, kclusters))
rainbow = [colors.rgb2hex(i) for i in colors_array]
#print(rainbow)
# add markers to the map

markers_colors = []

for lat, lon, nei , cluster in zip(borough_merged['Latitude'], 
                                   borough_merged['Longitude'], 
                                   borough_merged['Neighborhood'], 
                                   borough_merged['Cluster Labels']):
    label = folium.Popup(str(nei) + ' Cluster ' + str(cluster), parse_html=True)
    if not np.isnan(cluster):
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[int(cluster)-1],
            fill=True,
            fill_color=rainbow[int(cluster)-1],
            fill_opacity=0.7).add_to(map_clusters)

map_clusters
```
